```python
from OpenSMOG.OpenSMOG import SBM
```

```python
sbm_CA = SBM(name='1hk0', dt = 0.002, gamma = 5.0, rcutoff = 1.5, temperature = 0.7)
```

```python
sbm_CA.setup_openmm(platform='cuda', precision='mixed', GPUindex='default', integrator="langevin")
```

```python
sbm_CA.saveFolder('/home/antonio/smog2/1hk0/openSMOG/AA/output')
```

```python
sbm_CA.loadGro('/home/antonio/smog2/1hk0/openSMOG/AA/1hk0_AA.gro')
```

```python
sbm_CA.loadTop('/home/antonio/smog2/1hk0/openSMOG/AA/1hk0_AA.top')
```

```python
sbm_CA.loadXml('/home/antonio/smog2/1hk0/openSMOG/AA/1hk0_AA.xml')
```

```python
sbm_CA.createSimulation()
```

```python
sbm_CA.createReporters(trajectory=True, energies=True, dt_files=5*10**6)
```

```python
sbm_CA.run(5*10**8, report=True, interval=100)
```
